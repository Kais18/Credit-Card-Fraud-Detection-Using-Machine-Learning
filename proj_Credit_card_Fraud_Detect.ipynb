lets import the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to pandas
creditcard_data = pd.read_csv('/content/creditcsv.csv')
```

```python
creditcard_data.head()
#first 5 rows of data
```

```python
creditcard_data.tail()
```

```python
#legit and fraud classification
creditcard_data['Class'].value_counts()
```

```python
#classifying both
legit= creditcard_data[creditcard_data.Class == 0]
fraud= creditcard_data[creditcard_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#stats measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
creditcard_data.groupby('Class').mean()
#comapring both tnx
```

```python
#undersampling
#sample ds containing similar distribution of normal tnx and fraud tnx
legit_sample = legit.sample( n= 25 )
```

```python
#concatinaiting 2 Data Frames
new_dataset= pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
new_dataset.tail()

```

```python
new_dataset.groupby('Class').mean()
```

splitting the data into features and target

```python
X = new_dataset.drop(columns = 'Class', axis =1)
Y = new_dataset['Class']
print(X)
```

```python
print(Y)
```

split the data into Training data and testing data

```python
X_train, X_test ,Y_train, Y_test = train_test_split(X ,Y, test_size = 0.2, stratify= Y , random_state=2 )
```

```python
print(X.shape , X_train.shape, X_test.shape)
```

model Training
Logistic Regression

```python

model = LogisticRegression(max_iter=500)
model.fit(X_train , Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
